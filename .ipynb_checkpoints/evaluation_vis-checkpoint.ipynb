```python
def pie(stats, title):
    x = [len(i) for i in stats.values()]
    print(x)
    labels = stats.keys()
    plt.title(f"{title} {sum(x)} runs")
    plt.pie(x, autopct='%1.1f%%', labels=labels);
```

```python
def times(stats, title):
    # x = [sum(i)/len(i) for i in stats.values()]
    tmp = stats.copy()
    del tmp['timeout']
    labels = tmp.keys()
    plt.title(title)
    # plt.bar(range(len(x)), x);
    plt.boxplot(tmp.values())
    plt.xticks(range(1, len(x) + 1), labels)
    plt.ylabel("Average seconds")
```

```python
def times_compare(stats1, stats2, title):
    # x = [sum(i)/len(i) for i in stats.values()]
    def del_timeout(s):
        tmp = s.copy()
        del tmp['timeout']
        return tmp
    tmp1 = del_timeout(stats1)
    tmp2 = del_timeout(stats2)
    labels = tmp.keys()
    plt.title(title)
    tmp1['success']
    # plt.bar(range(len(x)), x);
    plt.boxplot(tmp1['success'])
    plt.xticks(range(1, len(x) + 1), labels)
    plt.ylabel("Average seconds")
```
